```python
from scipy import stats
import numpy as np
from matplotlib import pyplot as plt

SEED = 42
```

## Part 1

Generate 1000 random deviates from a B(20,.4) distribution, compute the sample mean and compare with the theoretical expectation.

```python
binom = stats.binom

num_samples = 1000
n = 20
p = 0.5

print(f"\nFor a binomial distribution with n = {n} and p = {p}:\n")

rvs = binom.rvs(n=n, p=p, size=num_samples, random_state=SEED)

print(f"Mean of {num_samples} samples: {np.mean(rvs)}")

# Theoretical mean of binomial distribution E(X) = n * p [Theorem 3.5.1]
theoretical_mean = n * p

print(f"Theoretical mean = E(X) = n*p = {theoretical_mean}")
```

## Part 2

Generate 1000 random deviates from a hypergeometric distribution with 4 white balls, 8 black balls, and 3 balls drawn without replacement. Compute the sample mean and compare with the theoretical expectation.

```python
hypergeom = stats.hypergeom

num_samples = 1000

r = 7 # number of red balls
w = 3 # number of white balls
n = 5 # number of draws

print(f"\nFor a hypergeometric distribution with M = {r+w}, n = {r}, and N = {n}:\n")

rvs = hypergeom.rvs(M=(r+w), n=r, N=n, size=num_samples, random_state=SEED)

print(f"Mean of {num_samples} samples: {np.mean(rvs)}")

# Theoretical mean of hypergeometric distribution E(X) = rn /(r+w) [Theorem 3.5.2]
theoretical_mean = (n*r) / (r + w)

print(f"Theoretical mean = E(X) = nr/(r+w) = {theoretical_mean}")
```

## Part 3

Generate 1000 random deviates from an exponential with lambda=3. Compute the sample mean and compare with the theoretical expectation,
which is 1/lambda.

```python
expon = stats.expon

num_samples = 1000
lambda_ = 3
scale = 1 / lambda_

print(f"\nFor an exponential distribution with lambda = {lambda_}:\n")

rvs = expon.rvs(scale=scale, size=num_samples, random_state=SEED)

print(f"Mean of {num_samples} samples: {np.mean(rvs)}")

# Theoretical mean of exponential distribution E(X) = 1/lambda
theoretical_mean = 1 / lambda_

print(f"Theoretical mean = E(X) = 1/lambda = {theoretical_mean}")
```

## Part 4

Generate the following exponential densities over the range (0.1, 10):
- lambda = 4
- lambda = 2
- lambda = 1
- lambda = 0.5

```python
lambdas = [4,2,1,0.5]
colors = ['black', 'red', 'blue', 'green']

min_x = 0.1
max_x = 10

x = np.linspace(min_x, max_x, 1000)

plt.figure(figsize=(6, 3))

for lambda_, color in zip(lambdas, colors):
    scale = 1 / lambda_
    plt.plot(x, lambda_*np.exp(-lambda_*x), label=f'lambda={lambda_}', color=color)

plt.legend()
plt.title('Exponential Distribution PDFs for Different Lambda Values')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid()
```
